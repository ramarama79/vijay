```python
import numpy as np
import matplotlib.pyplot as plot
from sklearn import svm
```

```python
rng=np.random.RandomState(0)
n_samples_1=1000
n_samples_2=100
```

```python
X=np.r_[1.5 * rng.randn(n_samples_1,2),
        0.5 * rng.randn(n_samples_2,2)+[2,2]]
```

```python
y=[0]*(n_samples_1)+[1]*(n_samples_2)
```

```python
clf=svm.SVC(kernel='linear',C=1.0)
```

```python
clf.fit(X,y)
```

```python
w=clf.coef_[0]
```

```python
a=-w[0]/w[1]
```

```python
xx=np.linspace(-5,5)
```

```python
yy=a*xx-clf.intercept_[0]/w[1]
```

```python
wclf=svm.SVC(kernel='linear',class_weight={1:10})
```

```python
wclf.fit(X,y)
```

```python
ww=wclf.coef_[0]
```

```python
wa=-ww[0]/ww[1]
```

```python
wyy=wa*xx-wclf.intercept_[0]/ww[1]
```

```python
h0=plt.plot(xx,yy,'k-',label='no weights')
```

```python
h1=plt.plot(xx,wyy,'k--',label='withweights')
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.Paired)
```

```python
plt.legend()
```

```python
plt.axis('tight')
```

```python
plt.show()
```

```python
print(__doc__)

import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm


def plot_decision_function(classifier, sample_weight, axis, title):
     xx, yy = np.meshgrid(np.linspace(-4, 5, 500), np.linspace(-4, 5, 500))
     
     Z = classifier.decision_function(np.c_[xx.ravel(), yy.ravel()])
     Z=Z.reshape(xx.shape)
     axis.contourf(xx,yy,Z,alpha=0.75,cmap=plt.cm.bone)
     axis.scatter(X[:,0],X[:,1],c=y,s=100*sample_weight,alpha=0.9,cmap=plt.cm.bone)
     axis.axis('off')
     axis.set_title(title)
```

```python
np.random.seed(0)
```

```python
X=np.r_[np.random.randn(10,2)+[1,1],np.random.randn(10,2)]
y=[1]*10+[-1]*10
```

```python
sample_weight_last_ten=abs(np.random.randn(len(X)))
sample_weight_constant=np.ones(len(X))
sample_weight_last_ten[15:]*=5
sample_weight_last_ten[9]*=15
```

```python
clf_weights=svm.SVC()
clf_weights.fit(X,y,sample_weight=sample_weight_last_ten)
```

```python
clf_no_weights=svm.SVC()
clf_no_weights.fit(X,y)
```

```python
fig,axes=plt.subplots(1,2,figsize=(14,6))
```

```python
plot_decision_function(clf_no_weights,sample_weight_constant,axes[0],"constantweights")
plot_decision_function(clf_weights,sample_weight_last_ten,axes[1],'Modifiedweights')
```

```python
plt.show()
```
