```python
var=boston.groupby('crim').medv.mean().reset_index()
```

```python
plt.plot(var.crim,var.medv)
plt.xlabel("crim")
plt.ylabel("mean of medv")
plt.title("mean of medv crim wise from boston")
```

```python
plt.show()
```

The crime rate is very less on high income class compared to low income class. 

```python
plt.bar(var.crim,var.medv,label='bar1',color='b')
plt.xlabel("crim rate per town")
plt.ylabel("median value of owner occupied homes(in 1000$)")
plt.title("crim rate compared to median income of household")
```
